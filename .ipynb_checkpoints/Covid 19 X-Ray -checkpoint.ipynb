```python
import numpy as np
import pandas as pd
import keras
import tensorflow as tf
from keras.layers import Dense
from keras.models import Sequential
import matplotlib.pyplot as plt

from keras.preprocessing.image import ImageDataGenerator
from keras.models import Model
from keras.layers import Input, Dropout, Activation, Conv2D, Flatten, MaxPooling2D
from sklearn.model_selection import train_test_split
from PIL import Image
%matplotlib inline
```

```python
df = pd.read_csv('Chest_xray_Corona_Metadata.csv')
df.tail()
```

```python
df.drop(['Unnamed: 0','Label', 'Label_2_Virus_category', 'Label_1_Virus_category'], axis = 1)
```

```python
X_train = df[df['Dataset_type'] == 'TRAIN']
X_test = df[df['Dataset_type'] == 'TEST']
```

```python
X_train.drop(['Unnamed: 0','Dataset_type', 'Label_2_Virus_category', 'Label_1_Virus_category'], axis = 1, inplace = True)
X_test.drop(['Unnamed: 0','Dataset_type', 'Label_2_Virus_category', 'Label_1_Virus_category'], axis = 1, inplace = True)
X_train.head()
```

```python
TRAIN_FOLDER = "Coronahack-Chest-XRay-Dataset/train/"
TEST_FOLDER = "Coronahack-Chest-XRay-Dataset/test/"
```

```python
image = plt.imread(TRAIN_FOLDER + X_train['X_ray_image_name'].iloc[0])
plt.imshow(image)
print(image.size)
```

```python
X_train['Label'] = X_train['Label'].map({'Pnemonia': 'positive', 'Normal': 'negative'})
X_test['Label'] = X_test['Label'].map({'Pnemonia': 'positive', 'Normal':  'negative'})
X_train.head()
```

```python
model = Sequential([
    Conv2D(64, (3, 3), input_shape=(224, 224, 3), activation='relu'),
    MaxPooling2D((3, 3)),
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D((3, 3)),
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D((3, 3)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.4),
    Dense(32, activation='relu'),
    Dropout(0.4),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer= 'Adam', loss= 'binary_crossentropy',
             metrics=['accuracy'])
```

```python
train_image_generator = ImageDataGenerator(
    rescale=1./255,
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=90,
    width_shift_range=0.15,
    height_shift_range=0.15,
    horizontal_flip=True,
    zoom_range=[0.9, 1.25],
    brightness_range=[0.5, 1.5]
)
```

```python
X_finaltrain = train_image_generator.flow_from_dataframe(
    dataframe= X_train,
    directory=TRAIN_FOLDER,
    x_col='X_ray_image_name',
    y_col='Label',
    target_size=(224, 224),
    batch_size=8,
    shuffle=True,
    class_mode='binary'
)

X_finaltest = train_image_generator.flow_from_dataframe(
    dataframe= X_test,
    directory=TEST_FOLDER,
    x_col='X_ray_image_name',
    y_col='Label',
    target_size=(224, 224),
    batch_size=8,
    shuffle=True,
    class_mode='binary'
)
```

```python
class AccuracyHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.acc = []

    def on_epoch_end(self, batch, logs={}):
        if logs.get('accuracy') > 0.99:
                print(f'Accuracy reached {logs.get("accuracy")*100:0.2f}. Stopping the training')
                self.model.stop_training = True

history = AccuracyHistory()
```

```python
model.fit(X_finaltrain, epochs = 50, callbacks = [history])
```

```python
model.save("PneumoniaDetector.h5")
```

```python
preds = model.predict(X_finaltest)
```
